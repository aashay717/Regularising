# **Occluded Shape Reconstruction Using OpenCV and Contour Detection**

This notebook demonstrates how to reconstruct an occluded curve by fitting it to a known geometric form and extrapolating the missing part. We'll use OpenCV for contour detection and geometric fitting, and visualize the results at each step.

### Steps:
1. Load and enhance the occluded image for contour detection.
2. Detect and separate the two main contours (occluded curve and occluding shape).
3. Fit the visible part of the curve to a regular shape close to it.
4. Extrapolate the missing part of the curve using the fitted ellipse.
5. Generate the final output by merging the reconstructed curve with the occluding shape.

## Step 2: Detect and Separate the Two Shapes

In this step, we'll detect the contours of the two main shapes in the image. These include the visible part of the occluded curve and the shape that is causing the occlusion. We'll separate these two contours for further processing.


```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Function to detect contours and separate the two shapes
def detect_and_separate_shapes(image):
    contours, _ = cv2.findContours(image, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    contours = sorted(contours, key=cv2.contourArea, reverse=True)
    if len(contours) >= 2:
        curve_contour = contours[1]  # The second largest contour is likely the interrupted curve
        occluding_shape_contour = contours[0]  # The largest contour is likely the occluding shape
        return curve_contour, occluding_shape_contour
    return None, None

# Detect and separate the curves
curve_contour, occluding_shape_contour = detect_and_separate_shapes(occlusion_image)

# Visualize the contours
contour_image = cv2.cvtColor(occlusion_image, cv2.COLOR_GRAY2BGR)
cv2.drawContours(contour_image, [curve_contour], -1, (0, 255, 0), 2)  # Green for curve contour
cv2.drawContours(contour_image, [occluding_shape_contour], -1, (255, 0, 0), 2)  # Blue for occluding shape contour

plt.figure(figsize=(6, 6))
plt.imshow(contour_image)
plt.title('Detected Contours')
plt.axis('off')
plt.show()
```

## Step 3: Fit the Visible Part of the Curve to a Regular Geometric Form

Next, we'll fit the detected curve contour to an ellipse. This will help us in reconstructing the missing part of the curve.


```python

def fit_curve_to_regular_form(curve_contour):
    if len(curve_contour) >= 5:
        curve_contour = curve_contour.astype(np.float32)
        ellipse = cv2.fitEllipse(curve_contour)
        return ellipse
    return None


fitted_ellipse = fit_curve_to_regular_form(curve_contour)


fitted_image = cv2.cvtColor(occlusion_image, cv2.COLOR_GRAY2BGR)
cv2.ellipse(fitted_image, fitted_ellipse, (0, 255, 0), 2)  # Green ellipse

plt.figure(figsize=(6, 6))
plt.imshow(fitted_image)
plt.title('Fitted Ellipse')
plt.axis('off')
plt.show()


# Function to convert ellipse parameters into contour points
def ellipse_to_points(ellipse_params, num_points=100):
    center, axes, angle = ellipse_params
    angle_rad = np.deg2rad(angle)
    points = []
    for i in range(num_points):
        theta = 2.0 * np.pi * i / num_points
        x = center[0] + axes[0] * np.cos(theta) * np.cos(angle_rad) - axes[1] * np.sin(theta) * np.sin(angle_rad)
        y = center[1] + axes[0] * np.cos(theta) * np.sin(angle_rad) + axes[1] * np.sin(theta) * np.cos(angle_rad)
        points.append([x, y])
    return np.array(points)

# Function to extrapolate the missing part based on the fitted regular form
def extrapolate_missing_curve(ellipse_params, curve_contour, occluding_shape_contour):
    ellipse_points = ellipse_to_points(ellipse_params)

    mask = np.zeros_like(occlusion_image, dtype=np.uint8)
    cv2.drawContours(mask, [occluding_shape_contour], -1, 255, thickness=cv2.FILLED)

    reconstructed_curve = []
    for point in ellipse_points:
        if mask[int(point[1]), int(point[0])] == 0:
            reconstructed_curve.append(point)

    reconstructed_curve = np.array(reconstructed_curve)
    return reconstructed_curve

# Extrapolate the missing part of the curve
if fitted_ellipse is not None:
    final_reconstructed_curve = extrapolate_missing_curve(fitted_ellipse, curve_contour, occluding_shape_contour)

    # Visualize the reconstructed curve
    reconstructed_image = cv2.cvtColor(occlusion_image, cv2.COLOR_GRAY2BGR)
    cv2.polylines(reconstructed_image, [final_reconstructed_curve.astype(np.int32)], isClosed=False, color=(0, 255, 0), thickness=2)

    plt.figure(figsize=(6, 6))
    plt.imshow(reconstructed_image)
    plt.title('Reconstructed Curve')
    plt.axis('off')
    plt.show()
else:
    print("Failed to fit the curve to a regular geometric form.")


# Function to generate the final output by filling the occluded curve and excluding the occluding curve
def generate_final_output(curve_contour, reconstructed_curve, occluding_shape_contour, image_size):
    final_image = np.zeros((image_size[0], image_size[1], 3), dtype=np.uint8)

    cv2.drawContours(final_image, [occluding_shape_contour.astype(np.int32)], -1, (255, 255, 255), 2)  # White
    cv2.drawContours(final_image, [reconstructed_curve.astype(np.int32)], -1, (0, 255, 0), 2)  # Green

    return final_image

# Generate the final merged output image
final_merged_output_image = generate_final_output(curve_contour, final_reconstructed_curve, occluding_shape_contour, occlusion_image.shape)

# Display the final reconstructed image
plt.figure(figsize=(6, 6))
plt.imshow(final_merged_output_image)
plt.title('Final Reconstructed Image')
plt.axis('off')
plt.show()

# Optionally, save the final output image
cv2.imwrite('/content/final_merged_output_image.png', final_merged_output_image)
```
